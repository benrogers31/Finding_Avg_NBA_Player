```python
from bs4 import BeautifulSoup as soup  # HTML data structure
#This allows us to make requests 
from urllib.request import urlopen as uReq  # Web client
```

```python
for year in range(1950,2021,1):
    
    #Here we enter the URL we are trying to web scrape from 
    page_url = "https://www.basketball-reference.com/leagues/NBA_" + str(year) + "_per_game.html"

    #this basically downloads the webpage into an object
    uClient = uReq(page_url)
    # parses html into a soup data structure to traverse html
    # as if it were a json data type.
    page_soup = soup(uClient.read(), "html.parser")
    uClient.close()

    # Storing all player names and team names into an arrays
    name_containers = page_soup.findAll("td", {"class": "left"})
    names = []
    team_names = []
    isName = True
    for c in name_containers: 
        #the names rotate from player name to team name so the flag changes each time
        if isName:
            names.append(c.text)
            isName = False 
        else:
            team_names.append(c.text)
            isName = True

    #now we are going to get the positions of each player
    position_containers = page_soup.findAll("td", {"class": "center"})
    positions = []

    for c in position_containers:
        positions.append(c.text)

    #now we are going to get rest of stats 
    stat_containers = page_soup.findAll("td", {"class": "right"})

    #### these are all of the stats we are going to store #### 
    ages = []
    games_played = []
    games_started = []
    minutes_played = []
    fg_made = []
    fg_attempts = []
    fg_percent = []
    three_pt_made = []
    three_pt_attempts = []
    three_pt_percent = []
    two_pt_made = []
    two_pt_attempts = []
    two_pt_percent = []
    eFG_percent = []
    ft_made = []
    ft_attempts = []
    ft_percent = []
    orb = []
    drb =[]
    trb = []
    ast =[]
    stl=[]
    blk= []
    tov = []
    pf =[]
    pts=[]


    for i,c in enumerate(stat_containers,start=0 ):
        #gets age 
        if(i%26 == 0):
            ages.append(c.text)
        #gets games played
        if(i%26 == 1):
            games_played.append(c.text)
        #gets games started 
        if(i%26 == 2):
            games_started.append(c.text)
        #gets minutes played
        if(i%26 == 3):
            minutes_played.append(c.text)
        #gets fg_made
        if(i%26 == 4):
            fg_made.append(c.text)
        #gets fg_attempts
        if(i%26 == 5):
            fg_attempts.append(c.text)
        #gets fg_percent
        if(i%26 == 6):
            fg_percent.append(c.text)
        #gets three_pt_made
        if(i%26 == 7):
            three_pt_made.append(c.text)
        #gets three_pt_atempts
        if(i%26 == 8):
            three_pt_attempts.append(c.text)
        #gets three_pt_percent
        if(i%26 == 9):
            three_pt_percent.append(c.text)
        #gets two_pt_made
        if(i%26 == 10):
            two_pt_made.append(c.text)
        #gets two_pt_attempts
        if(i%26 == 11):
            two_pt_attempts.append(c.text)
        #two_pt_percent
        if(i%26 == 12):
            two_pt_percent.append(c.text)
        #eFG_percent
        if(i%26 == 13):
            eFG_percent.append(c.text)
        #ft_made
        if(i%26 == 14):
            ft_made.append(c.text)
        #ft_attempts
        if(i%26 == 15):
            ft_attempts.append(c.text)
        #ft_percent
        if(i%26 == 16):
            ft_percent.append(c.text)
        #orb
        if(i%26 == 17):
            orb.append(c.text)
        #drb
        if(i%26 == 18):
            drb.append(c.text)
        #trb
        if(i%26 == 19):
            trb.append(c.text)
        #ast
        if(i%26 == 20):
            ast.append(c.text)
        #stl 
        if(i%26 == 21):
            stl.append(c.text)
        #blk
        if(i%26 == 22):
            blk.append(c.text)
        #tov
        if(i%26 == 23):
            tov.append(c.text)
        #pf
        if(i%26 == 24):
            pf.append(c.text)
        if(i%26 == 25):
            pts.append(c.text)

    out_filename = "csv_files/player_stats_"+str(year)+".csv"
    headers = "NAME,TEAM,POS,AGE,G,GS,MP,FG,FGA,FG%,3P,3PA,3P%,2P,2PA,2P%,eFG%,FT,FTA,FT%,ORB,DRB,TRB,AST,STL,BLK,TOV,PF,PTS\n"
    f = open(out_filename, "w")
    f.write(headers)

    for i,name in enumerate(names,start=0):

        f.write(names[i]+ ", " + team_names[i] + ", " + positions[i] + "," + ages[i] + "," + games_played[i]+ ", " + games_started[i] + ", " + minutes_played[i] + "," + fg_made[i] + "," + fg_attempts[i]+ ", " + fg_percent[i] + ", " + three_pt_made[i] + "," + three_pt_attempts[i] + "," + three_pt_percent [i]+ ", " + two_pt_made[i] + ", " + two_pt_attempts[i] + "," +  two_pt_percent[i] + "," +  eFG_percent[i] + ", " + ft_made[i] + "," +  ft_attempts[i] + "," + ft_percent[i]+ ", " + orb[i] + "," + drb[i] + "," +  trb[i] + "," + ast[i]+ ", " + stl[i] + ", " + blk[i] + "," + tov[i] + "," + pf[i]+ ", " + pts[i] +"\n")

    f.close()
    
  


```
